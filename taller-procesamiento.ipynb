```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargamos los datos de los dos datset para unirlos mediante el Id y proceder  realizar estadisticas iniciales.

```python
dataset1 = pd.read_excel('DatosP01.xlsx')
dataset2 = pd.read_excel('DatosP02.xlsx')
```

Unimos los data set con la función merge mediante el Id

```python
dataset = pd.merge(dataset1, dataset2[['Id', 'Tipo de Colegio', 'Uso_LMS', 'Asistencia', 'Entregas_Completas', 'Sistemas', 'Matematicas', 'Ingles', 'Comunicación']])
```

### Empezamos con unas estadisticas iniciales para saber como estamos con la información

```python
dataset.describe()
```

```python
dataset.info()
```

```python
# Calculamos los nulos que existen en el dataset completo
dataset.isnull().sum()
```

### Cambiamos los nombres de las columnas para poder manejarlas mejor

Cambiamos los nombres de las columnas por nombres en minusculas y sin espacios, nos ayudará a manejar de una manera más eficiente y rápida.

```python
dataset.columns = ['id', 'nombre', 'sexo', 'fecha_nacimiento', 'estrato', 'trabaja', 'enfermedad', 'ubicacion', 'tipo_colegio', 'uso_lms', 'asistencia', 'entregas_completas', 'sistemas', 'matematicas', 'ingles', 'comunicacion']
```

### Aqui vamos a corregir los valores de sexo. Dejarlo con femeninos y masculinos

Lo primero que vamos a realizar es cambiar en la columna `sexo` los campos que contengan `F` y `Mujer` por `Femenino`

```python
dataset.loc[(dataset['sexo'] == 'F') | (dataset['sexo'] == 'Mujer'), ['sexo']] = 'Femenino'
```

Luego vamos hacer lo mismo pero para los campos de index 19 y 20 debido a que son Mujeres pero aparecen con `M` y son los unicos dos registros con esa caracteristicas.

```python
dataset.loc[(dataset.index == 19) | (dataset.index == 20), ['sexo']] = 'Femenino'
```

Ahora si, realizamos la primera converción pero para los hombres. Se sustituyen los valores `M` por `Masculino`

```python
dataset.loc[(dataset['sexo'] == 'M'), ['sexo']] = 'Masculino'
```

De esta forma ya tendremos una columna `sexo` mucho más limpia que antes.

```python
dataset
```

### Acá vamos a establecer SI y NO en la columna de `trabaja`.

Traemos todos los campos que en la serie de trabaja que son iguales a `no` y `NO` y los dejamos en `NO`

```python
dataset.loc[(dataset['trabaja'] == 'NO') | (dataset['trabaja'] == 'no'), ['trabaja']] = 'NO'
```

Para los campos que aparecen en o en el dataset lo que vamos hacer es obtener la moda y remplazamos.

* Obtenemos la moda mediante la funcion mode() y de ella obtenemos los valores en la posición 0

```python
moda = dataset['trabaja'].mode().values[0]
dataset.loc[(dataset['trabaja'] == 0), ['trabaja']] = moda
```

```python
dataset
```

### Ahora vamos a corregir los de enfermedad

Realizamos el mismo proceso que con la columna `trabaja`

```python
dataset.loc[(dataset['enfermedad'] == 'NO') | (dataset['enfermedad'] == 'no'), ['enfermedad']] = 'NO'
```

```python
# Obtenemos la media de las enfermedades para no elilimnar la info:
moda_enfer = dataset['enfermedad'].mode().values[0]
```

```python
dataset.loc[(dataset['enfermedad'] == 0), ['enfermedad']] = moda_enfer
```

```python
dataset
```

### Con la columna de `ubicacion` vamos a remplazar los valores fltantres por la moda

```python
dataset.loc[(dataset['ubicacion'] == 0), ['ubicacion']] = dataset['ubicacion'].mode().values[0]
```

```python
dataset
```

Y el tipo ### Con la columna de `tipo__colegio` vamos a remplazar los valores fltantres por la moda

```python
dataset.loc[(dataset['tipo_colegio'] == 0), ['tipo_colegio']] = dataset['tipo_colegio'].mode().values[0]
```

```python
dataset
```

### Organizamos la columna de `uso_lms` con el mismo trabajo hecho en la columna `sexo`

```python
dataset.loc[(dataset['uso_lms'] == 'BAJA') | (dataset['uso_lms'] == 'baja'), ['uso_lms']] = 'BAJA'
```

```python
dataset.loc[(dataset['uso_lms'] == 'MEDIA') | (dataset['uso_lms'] == 'media'), ['uso_lms']] = 'MEDIA'
```

```python
dataset.loc[(dataset['uso_lms'] == 'ALTA') | (dataset['uso_lms'] == 'alta'), ['uso_lms']] = 'ALTA'
```

```python
dataset.loc[(dataset['uso_lms'] == 0), ['uso_lms']] = dataset['uso_lms'].mode().values[0]
```

```python
dataset
```

### Luego, remplazamos en la columna `entregas_completas` los campos vacios por `0`

```python
dataset.loc[dataset['entregas_completas'] == ' ', ['entregas_completas']] = 0
```

```python
dataset
```

### Por último, realizaremos una acumulado de las notas y sacaremos el estado del alumno.

* Lo primero será sumar los campos de `sistemas`, `metamaticas`, `ingles` y `comunicacion`

```python
dataset['acumulado'] = (dataset['sistemas'] + dataset['matematicas'] +  dataset['ingles'] +  dataset['comunicacion']) / 4
```

* Despues, crearemos una columna `Estado` que nos permitirá indicar si el alumno aprobó o no. Esto depende de la columna `àcumulado` si es mayor a 3.5 `APRUEBA` de lo contrario `NO APRUEBA`

```python
dataset.loc[(dataset['acumulado'] > 3.5), ['Estado']] = 'APROBADO'
```

El resto de los resultados los vamos a marcar como `NO APROBADO`

```python
dataset.loc[(dataset['acumulado'] <= 3.5), ['Estado']] = 'NO APROBADO'
```

```python
dataset
```

## Realizamos estadisticas de los datos ya listo
Graficamos los estado de cada alumna para tener conocimiento del porcentaje en diferencias de quienes aprobaron y quienes no

```python
fig, ax = plt.subplots()
estados = ['Aprobado', 'No Aprobado']
estados_count = dataset.groupby('Estado')['Estado'].count()
bar_colors = ['tab:green', 'tab:red']

ax.bar(estados, estados_count, label=estados, color=bar_colors)
for i, count in enumerate(estados_count):
    ax.text(i, count, f"{int((count/40)*100)} %", ha='center', va='bottom')
    
ax.set_ylabel('Cantidades')
ax.set_xlabel('Estados')
ax.set_title('Cantidad de Aprovados y No Aprobados')
plt.show()
```

Despues de haber planteado la grafica anterior, nos percatamos de que el 65% de los alumnos aprobaron satisfactoriamente. al contrario de los que no aprovaron que fue un total de 35%

```python
fig, ax = plt.subplots()
estados = ['NTNE', 'NTSE', 'STNE', 'STSE']
enfermos = dataset.groupby(['trabaja', 'enfermedad'])['enfermedad'].agg('count')
bar_colors = ['tab:pink', 'tab:blue', 'tab:red', 'tab:green']

ax.pie(enfermos, labels=estados, colors=bar_colors, autopct='%1.1f%%')
    
ax.set_ylabel('Cantidades')
ax.set_xlabel('Trabajadores')
ax.set_title('Cantidad de Trabajadores  que se encuentran enfermos')
ax.legend(labels=["No Tranaja No Enfermo", "No Trabaja Si Enfermo","Si Trabaja No Enfermo", "Si Trabaja Si Enfermo"])
plt.show()
```

Con esta gráfica nos percatamos de que el porcentaje de personas que no trabajan y que a su vez no estan enfermos (65%) es superior al porcentaje de los que si trabajan y si lo estan (5%). Además, que el porcentaje de los que Trabajan y no estan enfermos es practicamente igual al de los que no trabajan y si estan enfermos (15%). Hay una posibilidad de que esos alumnos que no trabajan y que no estan enfermos hayan aprobado el colegio.

```python
nea = dataset.loc[(dataset['trabaja'] == 'NO') & (dataset['enfermedad'] == 'NO') & (dataset['Estado'] == 'APROBADO')]['Estado'].count()
nen = dataset.loc[(dataset['trabaja'] == 'NO') & (dataset['enfermedad'] == 'NO') & (dataset['Estado'] == 'NO APROBADO')]['Estado'].count()
total = dataset.loc[(dataset['trabaja'] == 'NO') & (dataset['enfermedad'] == 'NO')]['Estado'].count()
```

```python
fig, ax = plt.subplots()
estados = ['NEA', 'SEA']
estados_count = [nea, nen]
bar_colors = ['tab:green', 'tab:red']

ax.bar(estados, estados_count, label=estados, color=bar_colors)
for i, count in enumerate(estados_count):
    ax.text(i, count, f"{int((count/total)*100)} %", ha='center', va='bottom')
    
ax.set_ylabel('Cantidades')
ax.set_xlabel('Estados')
ax.set_title('Cantidad de Aprovados y No Aprobados')
ax.legend(labels=["No Trabaja No Enfermo Aprobados", "No Trabaja No Enfermo No Aprobados"])
plt.show()
```

Como lo habiamos planteado en la estadistica anterior, nos podemos dar cuenta de que más del 50% de los alumnos que no trabajan y que no estan enfermos aprobaron el colegio, en comparación de los que no trabajan y que no estan enfermos pero que no aprobaron (34%).
